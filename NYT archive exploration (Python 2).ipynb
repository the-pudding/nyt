```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
import re
from nltk.stem.porter import PorterStemmer
porter_stemmer = PorterStemmer()
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
 


def stemming_tokenizer(str_input):
    words = re.sub(r"[^A-Za-z0-9\-]", " ", str_input).lower().split()
#     words = [porter_stemmer.stem(word) for word in words]
    words = [wordnet_lemmatizer.lemmatize(word) for word in words]
    return words
```

```python
count_vectorizer = CountVectorizer(stop_words='english', tokenizer=stemming_tokenizer)
X = count_vectorizer.fit_transform(yearly_df.iloc[9,1])
# print(count_vectorizer.get_feature_names())
feb_word_count = X.toarray().sum(axis=0)

word_list=[]

# for word in range(0,len(count_vectorizer.get_feature_names())):
for word_name, word_occurence in zip(count_vectorizer.get_feature_names(), feb_word_count):
    word_pair={}
    word_pair['word'] = word_name
    word_pair['occurences'] = word_occurence
    word_list.append(word_pair)
    del word_pair

df_1900_count = pd.DataFrame(word_list)

df_1900_count.sum()
```

```python
date_cut_word_occurrences = pd.DataFrame(df_1900_count.sum()).iloc[0,0]

df_1900_count['frequency']= df_1900_count['occurences']/float(date_cut_word_occurrences)
```

```python
from functools import reduce

all_words_list = list(yearly_df['para'])
flat_list = reduce(lambda x,y: x+y,all_words_list)
```

```python
count_vectorizer = CountVectorizer(stop_words='english', tokenizer=stemming_tokenizer)
X = count_vectorizer.fit_transform(flat_list)
# print(count_vectorizer.get_feature_names())
all_word_count = X.toarray().sum(axis=0)

word_list=[]

# for word in range(0,len(count_vectorizer.get_feature_names())):
for word_name, word_occurence in zip(count_vectorizer.get_feature_names(), all_word_count):
    word_pair={}
    word_pair['word'] = word_name
    word_pair['occurences'] = word_occurence
    word_list.append(word_pair)
    del word_pair

df_all_years = pd.DataFrame(word_list)

df_all_years.sum()
```

```python
date_cut_word_occurrences = pd.DataFrame(df_all_years.sum()).iloc[0,0]

df_all_years['frequency']= df_all_years['occurences']/float(date_cut_word_occurrences)
```
